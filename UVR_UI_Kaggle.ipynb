# Run UI
Select the type of tunnel you wanna use for seeing the public link, so that if one of them is down, you can use the other one:

A) Ngrok: Select it in Tunnel, get the Ngrok Authtoken here: https://dashboard.ngrok.com/tunnels/authtokens/new, put it in ngrok_authtoken, you can optinally change the Ngrok Tunnel Region to one nearer to you for lower latency, run the cell, wait for the Local URL to appear and click on the Ngrok Tunnel Public URL which is above.

B) LocalTunnel: Select it in Tunnel, run the cell, wait for the Local URL to appear, copy the LocalTunnel Password displayed under the public link and paste it in Tunnel Password of the LocalTunnel Tunnel Public URL which is above.

C) Horizon: Select it in Tunnel, get the Horizon ID here: https://hrzn.run/dashboard/ , login, on the 2nd step, it shows an hrzn login YOUR_ID, you need to copy that id and put it in horizon_id. Then run the cell, you will get an 'HORIZON: Authorize at https://hrzn.run/dashboard/settings/cli-token-requests/YOUR_ID', click it, do Approve. At the end, run the cell, wait for the Local URL to appear and click on the Horizon Tunnel Public URL which is above.

Also, you need to wait for the Local URL to appear, and only after that click on the Public URL which is above.

```python
import codecs
import os
path = codecs.decode('/xnttyr/jbexvat/HIE5-HV','rot_13')
%cd $path
app = codecs.decode('ncc.cl','rot_13')

Tunnel = "Ngrok" #@param ["Ngrok", "LocalTunnel", "Horizon"]

ngrok_authtoken = "" #@param {type:"string"}
ngrok_region = "us - United States (Ohio)" # @param ["au - Australia (Sydney)","eu - Europe (Frankfurt)", "ap - Asia/Pacific (Singapore)", "us - United States (Ohio)", "jp - Japan (Tokyo)", "in - India (Mumbai)","sa - South America (Sao Paulo)"]

horizon_id = "" #@param {type:"string"}

# tunnel management
if Tunnel == "Ngrok":
  from pyngrok import conf, ngrok
  NgrokConfig = conf.PyngrokConfig()
  NgrokConfig.auth_token = ngrok_authtoken
  NgrokConfig.region = ngrok_region[0:2]
  conf.set_default(NgrokConfig)
  ngrok.kill()
  main_tunnel = ngrok.connect(9999, bind_tls=True)
  file_tunnel = ngrok.connect(9876, bind_tls=True)
  print("UVR5 URL:", main_tunnel.public_url)
  print("File URL:", file_tunnel.public_url)
elif Tunnel == "LocalTunnel":
    # install
    import time
    import urllib
    # run localtunnel UVR5 Url
    with open('uvr5uiurl.txt', 'w') as file:
        file.write('')

    get_ipython().system_raw('lt --port 9999 >> uvr5uiurl.txt 2>&1 &')

    time.sleep(4)

    endpoint_ip = urllib.request.urlopen('https://ipv4.icanhazip.com').read().decode('utf8').strip("\n")

    with open('uvr5uiurl.txt', 'r') as file:
        tunnel_url = file.read()
        tunnel_url = tunnel_url.replace("your url is: ", "")

    clear_output()
    print(f"UVR5 UI Public URL: \033[0m\033[93m{tunnel_url}\033[0m", end="\033[0m")

    # run localtunnel File URL
    with open('fileurl.txt', 'w') as file:
        file.write('')

    get_ipython().system_raw('lt --port 9876 >> fileurl.txt 2>&1 &')

    time.sleep(4)

    with open('fileurl.txt', 'r') as file:
        tunnel_url = file.read()
        tunnel_url = tunnel_url.replace("your url is: ", "")

    print(f"File Public URL: \033[0m\033[93m{tunnel_url}\033[0m", end="\033[0m")
    
    print(f'LocalTunnels Password: {endpoint_ip}')
elif Tunnel == "Horizon":
  # install 
  !npm install -g @hrzn/cli
  import time
  # login
  !hrzn login $horizon_id
  # run horizon
  # rvc
  with open('url.txt', 'w') as file:
        file.write('')

  get_ipython().system_raw('hrzn tunnel http://localhost:9999 >> url.txt 2>&1 &')

  time.sleep(4)

  with open('url.txt', 'r') as file:
      tunnel_url = file.read()
      tunnel_url = !grep -oE "https://[a-zA-Z0-9.-]+\.hrzn\.run" url.txt
      tunnel_url = tunnel_url[0]

  clear_output()

  print(f"RVC Tunnel Public URL: \033[0m\033[93m{tunnel_url}\033[0m")

  # run horizon
  with open('url2.txt', 'w') as file:
        file.write('')

  get_ipython().system_raw('hrzn tunnel http://localhost:9876 >> url2.txt 2>&1 &')

  time.sleep(4)

  with open('url2.txt', 'r') as file:
      tunnel_url = file.read()
      tunnel_url = !grep -oE "https://[a-zA-Z0-9.-]+\.hrzn\.run" url.txt
      tunnel_url = tunnel_url[0]

  print(f"File Tunnel Public URL: \033[0m\033[93m{tunnel_url}\033[0m")

# kills previously running processes
!fuser -k 9999/tcp
!fuser -k 9876/tcp

# run imjoyelfinder
os.system(f". /kaggle/working/.venv/bin/activate; imjoy-elfinder --root-dir=/kaggle/working --port 9876 > /dev/null 2>&1 &")

# run UVR5 UI
!python $app
```
